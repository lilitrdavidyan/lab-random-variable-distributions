```python
import pandas as pd
numerical = pd.read_csv('numerical.csv')

numerical
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(20, 15))

for i, col in enumerate(numerical.columns, 1):
    plt.subplot(3, 3, i)
    sns.histplot(numerical[col], kde=True)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

## Observations:

- Variables like **income**, **monthly_premium_auto**, **number_of_open_complaints**, and **number_of_policies** appear to be right-skewed (positively skewed) rather than following a normal distribution.
- The distribution of **customer_lifetime_value** also appears to be right-skewed.
- The distributions of **months_since_last_claim** and **months_since_policy_inception** seem more evenly spread but still don't exhibit clear normal distribution characteristics.


```python
import numpy as np

# Applying log transformation to the numerical data + 1 (to avoid log(0))
numerical_log_transformed = np.log1p(numerical)

# Setting up the figure size for the plots
plt.figure(figsize=(20, 15))

# Iterating through each log-transformed numerical column to plot its distribution
for i, col in enumerate(numerical_log_transformed.columns, 1):
    plt.subplot(3, 3, i)
    sns.histplot(numerical_log_transformed[col], kde=True, color='skyblue')
    plt.title(f'Log-transformed Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

## Observations:

- The log transformation has made the distribution of variables like **customer_lifetime_value**, **income**, and **monthly_premium_auto** more bell-shaped, resembling a normal distribution.
- Variables such as **number_of_open_complaints** and **number_of_policies** still exhibit a certain level of skewness, but their distributions have been somewhat compressed towards the center compared to their original distributions.
- The distributions of **months_since_last_claim** and **months_since_policy_inception** have been slightly modified but still don't perfectly follow a normal distribution.



The following variables exhibit characteristics of an exponential distribution in their original form:

- `customer_lifetime_value`
- `income`
- `monthly_premium_auto`
- `number_of_open_complaints`
- `number_of_policies`

These variables display a rapid decrease in frequency as their values increase, which is indicative of an exponential distribution.
